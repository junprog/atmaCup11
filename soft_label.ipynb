```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.array([year for year in range(1400, 2000)])
```

```python
def soft_label(label):
    if 1801 <= label:
        soft = label * 0.002631578947 - 2.239473683

    elif 1600 < label and label < 1801:
        soft = label / 100. - 15.51
    
    elif label <= 1600:
        soft = label * 0.003105590062 - 4.472049689

    return soft
```

```python
y = np.array([soft_label(year) for year in x])

plt.plot(x, y)
plt.show()
```
